$$ f(x) = x^3 + 4x^2 - 10$$

Intervalo: $[1, 2] $

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
f = lambda x: x**3 +4*x**2 -10
```

```python
f(1)
```

```python
f(2)
```

```python
x = np.arange(0,3,0.001)
y = f(x)
```

```python
plt.plot(x,y)
plt.grid()
plt.show()
```
